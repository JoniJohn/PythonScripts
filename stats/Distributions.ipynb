Normal Distribution - Simulating Sales Under New Market Conditions

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Calculate new average amount
new_mean = 5000 * 1.2

# Calculate new standard deviation
new_sd = 2000 * 1.3

# Simulate 36 new sales
new_sales = norm.rvs(new_mean, new_sd, size=36)

# Create a historram to show
plt.hist(new_sales, bins=10, alpha=0.75, color='blue')
plt.title('Simulated Sales Under New Market Condition')
plt.xlabel('Sale Amount')
plt.ylabel('Frequency')
plt.show()
```
